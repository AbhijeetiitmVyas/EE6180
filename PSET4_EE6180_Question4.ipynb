```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def channel(x,alpha):
    return(x+np.random.laplace(0,1.0/alpha,x.shape))
```

```python
X=np.loadtxt('./drugs.txt',delimiter = ',')
```

```python
theta_hat = []
for i in range(10):
    theta_hat.append(np.mean(channel(X,2**(-i))))
```

```python
plt.figure(figsize=(10,10))
plt.yticks(np.arange(0,5,0.1))
plt.xticks(np.arange(1,10,1))
plt.plot(np.arange(1,11,1),theta_hat,label = 'Estimated Mean')
plt.plot(np.arange(1,11,1),np.ones(10)*np.mean(X),label='True Mean')
plt.xlabel('alpha=2^(-x)')
plt.ylabel('Estimated Theta')
plt.legend()
```

### We see that the approximation worsens as alpha falls(We know the upperbound to follow 1/alpha^2):

```python
for j in range(10):
    theta_hat = []
    for i in range(10):
        theta_hat.append(np.mean(channel(X,2**(-i))))
    plt.figure(figsize=(10,10))
    plt.yticks(np.arange(0,5,0.1))
    plt.xticks(np.arange(1,10,1))
    plt.plot(np.arange(1,11,1),theta_hat,label = 'Estimated Mean')
    plt.plot(np.arange(1,11,1),np.ones(10)*np.mean(X),label = 'True Mean')    
    plt.xlabel('alpha=2^(-x)')
    plt.ylabel('Estimated Theta')
    plt.legend()

```
